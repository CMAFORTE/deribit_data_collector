```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append("..")
import deribit_data as dm
import requests
```

```python
data = dm.Options("BTC")
```

```python
df = data.get_options_stats()
df['mid_iv'] = (df.bid_iv + df.ask_iv)/2
```

```python
call = df[df.option_type=='call']
put = df[df.option_type=='put']
```

## Greeks and Volatility Analysis:

### Analyzing the Volatility Smile 

```python
# add in a non-linear regression - ie: loess or KNN regression

# AAPL Call Options 
r = requests.get("https://query2.finance.yahoo.com/v7/finance/options/aapl?date=1590710400")
s = pd.DataFrame(r.json()['optionChain']['result'][0]['options'][0]['calls'])

# BTC Call Options 
temp = call.sort_values('strike')
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
aapl_kreg = KNeighborsRegressor()
aapl_kreg.fit(s[['strike']].values.reshape(-1,1), 
         s[['impliedVolatility']].values.reshape(-1,1))
```

```python
btc_kreg = KNeighborsRegressor()
btc_kreg.fit(temp[['strike']].values.reshape(-1,1), 
         temp[['mid_iv']].values.reshape(-1,1)/100)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axes[0].plot(s.strike, aapl_kreg.predict(s[['strike']].values.reshape(-1,1)), c='r', alpha=0.70)
axes[0].scatter(s.strike, s.impliedVolatility)
axes[0].set_title("AAPL Calls Volatility Smile")
axes[0].set_xlabel("Strike")
axes[0].set_ylabel("Implied Volatility")


axes[1].plot(temp.strike, btc_kreg.predict(temp[['strike']].values.reshape(-1,1)), c='r', alpha=0.70)
axes[1].scatter(temp.strike, temp.mid_iv/100)
axes[1].set_title("BTC Calls Volatility Smile?")
axes[1].set_xlabel("Strike")
axes[1].set_ylabel("Mid Implied Volatility")

fig.tight_layout()
```

### Analyzing the Volatility Term Strucure

```python
# BTC Options Near ATM 
atm_options = df[(df.strike == 9500.0)]
# consider all options - calls and puts
vol_structure = atm_options.groupby("expiration_timestamp").mean()
vol_structure.mid_iv.plot()
```

Upward sloping volatiltiy term-structure indicates the market is pricing in higher volatility in the future.
