```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append("..")
import deribit_data as dm
```

```python
eth_data = dm.Options("ETH")
```

```python
hist_vol = eth_data.get_hist_vol()
```

```python
# Plot past 15 days of historical annualised volatility for ETH 
hist_vol.plot()
```

```python
# Download all option chains for ETH
eth_options = eth_data.get_options_stats(save_csv=False)
```

```python
call = eth_options[eth_options.option_type=='call'].reset_index()
call_greeks = pd.DataFrame(call.greeks.tolist())
call_greeks['option_strike'] = call.strike
call_greeks.sort_values("option_strike", inplace=True)

put = eth_options[eth_options.option_type=='put'].reset_index()
put_greeks = pd.DataFrame(call.greeks.tolist())
put_greeks['option_strike'] = put.strike
put_greeks.sort_values("option_strike", inplace=True)
```

```python
plt.bar(call_greeks.option_strike, call_greeks.theta, alpha=0.10)
plt.title('Call Theta vs. Strike Price')
plt.show()
plt.bar(call_greeks.option_strike, call_greeks.vega, alpha=0.10)
plt.title('Call Vega vs. Strike Price')
plt.show()
plt.bar(call_greeks.option_strike, call_greeks.delta, alpha=0.10)
plt.title('Call Delta vs. Strike Price')
plt.show()
plt.bar(call_greeks.option_strike, call_greeks.gamma, alpha=0.10)
plt.title('Call Gamma vs. Strike Price');
```
